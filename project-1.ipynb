# Number guessing Game

```python
import random

sec_num = random.randint(1, 10)
attempts = 3
print("I'm thinking of a number between 1 and 10")

while attempts> 0:
    guess = int(input("Take a Guess: "))
    if guess == sec_num:
        print(f"You guessed it right, num is {sec_num}")
        break
    elif guess < sec_num:
        print("You guessed too low! Try again.")
        print("Your remaining attempts are: ", attempts-1)
    else:
        print("You guessed too high, try again")
        print("Your remaining attempts are: ", attempts-1)
    attempts -= 1
if attempts == 0:
    print(f"Sorry, you have run out of attempts. Secret num was: {sec_num}")
```

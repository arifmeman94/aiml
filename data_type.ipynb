```python
# Py version
import sys
print(sys.version)

# Data Types

# 1. Fundamental Data Types

# Int - Numeric value without any decimal
# Example: a=10
a = 10
print(a) # Output: 10

# To check data type: use type() 
print(type(a)) # Output: <class 'int'>
```

```python
# Float - Numeric value with decimal
# Example: b=3.0
b = 3.0
print(b) # Output: 3.0
print(type(b)) # Output: <class 'float'>
```

```python
# Type Casting: Convert from one data type to another data type
b = 5.9
print(int(b)) # Output: 5
print(int(10.9999)) # Output: 10
print(float(int(5.4))) # Output: 5.0
```

```python
# converting Int in to other data types
a = int(b)
print(a,type(a))
a = str(b)
print(a, type(a))
a = float(b)
print(a,type(a))
a = bool(b)
print(a, type(a))
```

```python
# converting float to other data types
b = 2.99
a = int(b)
print(a, type(a))
a= float(b)
print(a, type(a))
a= str(b)
print(a, type(a))
a = bool(b)
print(a, type(a))
```

```python
# Boolean
b = True
a = int(b)
print(a, type(a))
b= False
a = int(b)
print(a, type(a))
```

```python
# Complex
# Complex literals use the notation x+yj where x is the real component and y is the imaginary component.
x = 10 + 20.5j
print(type(x)) # Output: <class 'complex'>

# To get real value
print(x.real) # Output: 10.0

# To get imaginary value
print(x.imag) # Output: 20.5

# Cannot convert complex to int
# int(10+5j) # This raises TypeError: can't convert complex to int
print(complex(10.4)) # Output: (10.4+0j)
```

```python
# Strings
# A string in Python consists of a series or sequence of characters - letters, numbers, and special characters.
# We can use single quotes or double quotes to represent strings.

a = 'siva'
print(a) # Output: siva
print(type(a)) # Output: <class 'str'>

myString1 = 'Hello'
print(myString1) # Output: Hello

myString2 = "Hello"
print(myString2) # Output: Hello

myString3 = '''Hello'''
print(myString3) # Output: Hello
```

```python
# Multi-line strings can be denoted using triple quotes, or """
name = '''siva
rama
krishna'''
print(name) # Output: siva\nrama\nkrishna

var = "1234567"
print(type(var)) # Output: <class 'str'>
print(float(var)) # Output: 1234567.0
```

```python
# Length of string
myString = "Hello"
print(len(myString)) # Output: 5
```

```python
# Indexing (accessing characters in a string)
# The first character of a string has the index 0.
print(myString[0]) # Print first character. Output: H

# Negative indexing: -1 is the last character.
print(myString[-1]) # Output: o
print(myString[-3]) # Output: l

# If we try to access index out of range, we will get an IndexError.
# print(myString[5]) # This raises IndexError: string index out of range

# If we try to access by using use a decimal number, we will get a TypeError.
# print(myString[1.5]) # This raises TypeError: string indices must be integers
```

```python
# Slicing the strings
# [ start index (inclusive): end index (exclusive): increment/decrement ]
# [default increment =+1]

# Slicing 2nd to 5th character (start index to end-1 index)
print(myString[2:5]) # Output: 'llo' (indices 2, 3, 4)

print(myString[0:3]) # Output: 'Hel'

# Slicing (0 to end-1)
print(myString[:4]) # Output: 'Hell'

# Slicing with increment (index 0, skip 1, index 2, skip 3, index 4)
print(myString[0::2]) # Output: 'Hlo'

# Print in reverse order (using decrement)
print(myString[::-1]) # Output: 'olleH'
```

```python
# String Concatenation
# Joining of two or more strings into a single one is called concatenation.
# The + operator does this in Python.
print('Hello' + " World") # Output: 'Hello World'
print("hello " + "world") # Output: 'hello world'

# Can only concatenate str (not "int") to str
# print("siva"+2) # This raises TypeError

# The * operator can be used to repeat the string
print("AI" * 5) # Output: 'AIAIAIAIAI'
```

```python
# String Immutable
myString = "Hello"
print(myString[4]) # Output: 'o'

# Strings are immutable. This means that elements of a string cannot be changed once it has been assigned.
# myString[4]='s' # This raises TypeError: 'str' object does not support item assignment
```

```python
# String Methods
a = "gooD moRNing"

# .capitalize() --> Capitalizes only the first letter, and converts the rest to lowercase
print(a.capitalize()) # Output: 'Good morning'

# .title() --> Capitalizes the first letter of each word and converts the remaining all to small
print(a.title()) # Output: 'Good Morning'

# .upper() --> convert complete string to upper case
print(a.upper()) # Output: 'GOOD MORNING'

# .lower() --> convert complete string to lower case
print(a.lower()) # Output: 'good morning'

# .find(sub) --> Returns the index position of string if available, or -1 if not available
print(a.find("mo")) # Output: 5

# .count(sub) --> To count how many times the string available in the main string
print(a.count("o")) # Output: 3

# .split() --> Split with space (default)
st = 'my name is Sam'
print(st.split()) # Output: ['my', 'name', 'is', 'Sam']

# .replace(old, new) --> To replace the old string with new string
s1 = "Good morning"
s1 = s1.replace("Good", "Bad")
print(s1) # Output: Bad morning
```

```python
my_str = 'Siva123'

# .isdigit() # test if string contains only numbers
print(my_str.isdigit()) # Output: False

# .isalnum() # check if it is alphanumeric
print(my_str.isalnum()) # Output: True

# .isspace() # test if string contains spaces
print(my_str.isspace()) # Output: False

# .endswith('s') # test if string ends with a s
print(my_str.endswith('s')) # Output: False

# .startswith('S') # test if string starts with S
print(my_str.startswith('S')) # Output: True

# .istitle() # test if string contains title words
print(my_str.istitle()) # Output: False

# .isalpha() # check if all char in the string are alphabetic
print(my_str.isalpha()) # Output: False

# .isupper() # test if string contains upper case
print(my_str.isupper()) # Output: False

# .islower() # test if string contains lower case
print(my_str.islower()) # Output: False
```

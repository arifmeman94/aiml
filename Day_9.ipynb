```python
#Objects & Mutability
# Everythin is object in python
''' 
- every object have unique identity
- every object have type 

- Mutable : that is changable. thi can be changed.
- immutable : this can not be changed

- check identity, if it's changed then it is not muttable.
- if identity remains same, it is muttable


'''
```

```python
sugar_amount_sp = 2
print(f"initial sugar: {sugar_amount_sp}")
```

```python
id(sugar_amount_sp)
```

```python
sugar_amount_sp = 12
print(sugar_amount_sp)
```

```python
id(sugar_amount_sp)
```

```python
id(2)
```

```python
id(12)
```

```python
spice_mix = set()
print(f"initial spice mix id: {id(spice_mix)}")
```

```python
spice_mix.add("Ginger")
spice_mix.add("Cardamom")
print(f"After addind value: {id(spice_mix)}")
```

```python
num = 12
print(f"num:{id(num)}")
print(f"sug old: {id(sugar_amount_sp)}")
```
